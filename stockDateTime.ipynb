```python
import requests
import pandas as pd
import os


Alphavantage_API_KEY=os.environ.get('Alphavantage_API_KEY')
# https://www.alphavantage.co/query?function=MARKET_STATUS&apikey=demo
url = f"https://www.alphavantage.co/query?function=MARKET_STATUS&apikey={Alphavantage_API_KEY}"
response = requests.get(url)
data = response.json()
df = pd.DataFrame(data['markets'])
df
```

```python
priority_regions = ['United States', 'Mainland China', 'Hong Kong', 'Japan', 'South Korea', 'India']
priority_df = df[df['region'].isin(priority_regions)]
priority_df
```
